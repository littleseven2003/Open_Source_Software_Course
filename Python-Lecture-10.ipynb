```python
import inspect

x = 10

def outer():
    x = 20
    def inner():
        x = 30
        print(x)
        return inspect.currentframe()
    return inner

inn = outer()
frame = inn()
print(frame.f_locals['x'])
print(frame.f_globals['x'])
print(inspect.getclosurevars(inn))
```

```python
import inspect

x = 10

def outer():
    x = 20
    def inner():
        # x = 30
        print(x)  # 这里会打印 inner 的局部变量 x（如果定义了），否则会查找外层作用域
        return inspect.currentframe()
    return inner

inn = outer()
frame = inn()
print(frame.f_locals['x'])  # 打印 inner 函数的局部变量 x
print(frame.f_globals['x'])  # 打印全局变量 x
print(inspect.getclosurevars(inn))  # 打印 inner 函数的闭包变量信息
```

```python
from sys import settrace

def tracer(frame, event, arg=None):
    code = frame.f_code
    func_name = code.co_name
    line_no = frame.f_lineno
    print(f"事件：{event}，函数：{func_name}()，行号：{line_no}")
    return tracer

def foo():
    print("in foo")

def bar():
    foo()

old_tracer = settrace(tracer)
bar()
settrace(old_tracer)
```

```python
import tokenize
import io

code = '''\
def foo(bar):
    return bar + 1
'''

# 使用内存中的代码字符串
for token in tokenize.tokenize(io.BytesIO(code.encode('utf-8')).readline):
    print(token)

# 或者从文件中读取代码
with open('HelloWorld.py', 'rb') as f:
    for token in tokenize.tokenize(f.readline):
        print(token)
```

```python
import libcst

class FuncIntVisitor(libcst.CSTVisitor):
    def __init__(self):
        self.funcint = dict()
        self.current_func = None

    def visit_FunctionDef(self, node):
        name = node.name.value
        self.funcint[name] = 0
        self.current_func = name

    def visit_Integer(self, node):
        if self.current_func:
            self.funcint[self.current_func] += 1

    def leave_FunctionDef(self, node):
        print(self.current_func, self.funcint[self.current_func])
        self.current_func = None

text = '''\
def foo(bar):
    return 1

def bar():
    print('hello')

foo(3)
'''

code = libcst.parse_module(text)
visitor = FuncIntVisitor()
code.visit(visitor)
print(visitor.funcint)
```

```python
import libcst
import re

class FuncNameUpper(libcst.CSTTransformer):
    def leave_FunctionDef(self, node, updated):
        name = updated.name.value
        updated_name = re.sub(r'_([a-z])', lambda match: match.group(1).upper(), name)
        return libcst.FunctionDef(
            name=libcst.Name(value=updated_name),
            params=updated.params,
            body=updated.body
        )

text = '''\
def foo_foo_bar(bar):
    return 1

def bar():
    print('hello')

foo(3)
'''

code = libcst.parse_module(text)
transformer = FuncNameUpper()
updated = code.visit(transformer)
print(updated.code)
```
